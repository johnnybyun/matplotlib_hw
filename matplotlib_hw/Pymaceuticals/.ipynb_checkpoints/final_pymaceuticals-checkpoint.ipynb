```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load, encoding="ISO-8859-1")
clinical_trial_df = pd.read_csv(clinical_trial_data_to_load, encoding="ISO-8859-1")
```

```python
# Check for bad data in the mouse data set. Looks good. 
mouse_df.count()
```

```python
# Look for bad data in the clinical trial data set.  Looks good. 
clinical_trial_df.count()
```

```python
# Merge the two tables on 'Mouse ID' and look for any bad data.  
merge_table = pd.merge(mouse_df, clinical_trial_df, on="Mouse ID", how="left")
merge_table.count()
# The merge has increased the number of rows 1893 to 1906 which may indicate a duplicate.
```

```python
# We look for the duplicate. 
mouse_group = mouse_df.groupby(['Mouse ID'])
mouse_group.count().sort_values("Drug", ascending=False)
# g989 is associated with 2 drugs. 
```

```python
# Take a closer look at the g989 data.
g989 = merge_table.loc[merge_table["Mouse ID"]=="g989", :]
g989
# g989 is associated with two drugs.  
```

```python
# The 'g989' data is questionable so we drop it. 
new_merge = merge_table.drop(merge_table[(merge_table['Mouse ID'] == 'g989')].index)
new_merge.count()
# Confirm that the rows have been dropped. 
```

```python
new_merge.head()
dfa = new_merge['Mouse ID']
```

```python
# Rearrange the columns to match the output in the readme. 
merged_table = merge_table[['Mouse ID', 'Timepoint', 'Tumor Volume (mm3)', 'Metastatic Sites', 'Drug']]
merged_table.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
grouped_table_m = merge_table.groupby(['Drug', 'Timepoint']).mean()

# Convert to DataFrame
# mean_tumor_df = grouped_table_m.reset_index()[['Drug', 'Timepoint', 'Tumor Volume (mm3)']]
mean_tumor_df = grouped_table_m.reset_index()[['Drug', 'Timepoint', 'Tumor Volume (mm3)']]

# Preview DataFrame
mean_tumor_df
```

```python
# MEAN TUMOR:  Minor Data Munging to Re-Format the Mean Tumor Data Frame
new_mean_tumor_df = mean_tumor_df.pivot(index='Timepoint', columns='Drug',values='Tumor Volume (mm3)')
new_mean_tumor_df
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grouped_table_sem = merge_table.groupby(['Drug', 'Timepoint']).sem()
# Convert to DataFrame
sem_tumor_df = grouped_table_sem.reset_index()[['Drug', 'Timepoint', 'Tumor Volume (mm3)']]
# Preview DataFrame
sem_tumor_df
```

```python
# SEM: Minor Data Munging to Re-Format the SEM Data Frame
new_sem_tumor_df = sem_tumor_df.pivot(index='Timepoint', columns='Drug', values='Tumor Volume (mm3)')
new_sem_tumor_df
```

```python

# Generate the Plot 

plt.errorbar(new_mean_tumor_df.index, 
             new_mean_tumor_df['Capomulin'], 
             yerr=new_sem_tumor_df['Capomulin'], 
             marker='o', color='red', linewidth=0.5)

plt.errorbar(new_mean_tumor_df.index, 
             new_mean_tumor_df['Infubinol'], 
             yerr=new_sem_tumor_df['Infubinol'], 
             marker='o', color='blue', linewidth=0.5)

plt.errorbar(new_mean_tumor_df.index, 
             new_mean_tumor_df['Ketapril'], 
             yerr=new_sem_tumor_df['Ketapril'], 
             marker='o', color='green', linewidth=0.5)

plt.errorbar(new_mean_tumor_df.index, 
             new_mean_tumor_df['Placebo'], 
             yerr=new_sem_tumor_df['Placebo'], 
             marker='o', color='black', linewidth=0.5)

Capomulin, = plt.plot(new_mean_tumor_df.index, 
                     new_mean_tumor_df['Capomulin'], 
                     marker='o', color='red', label='Capomulin')

Infubinol, = plt.plot(new_mean_tumor_df.index, 
                     new_mean_tumor_df['Infubinol'], 
                     marker='o', color='blue', label='Infubinol')

Ketapril, = plt.plot(new_mean_tumor_df.index, 
                     new_mean_tumor_df['Ketapril'], 
                     marker='o', color='green', label='Ketapril')

Placebo, = plt.plot(new_mean_tumor_df.index, 
                     new_mean_tumor_df['Placebo'], 
                     marker='o', color='black', label='Placebo')

plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Response to Treatment')
plt.grid(axis='y')
plt.legend(loc='best')
plt.savefig('../Images/treatment.png')
plt.show()
```

```python
## METASTATIC SITE: Store the Mean Metastatic Site Data Grouped by Drug and Timepoint 
mean_metastatic_df = grouped_table_m.reset_index()[['Drug', 'Timepoint', 'Metastatic Sites']]
mean_metastatic_df
```

```python
new_mean_metastatic_df = mean_metastatic_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
new_mean_metastatic_df
```

```python
# Store the Standard Error of Metastatic Sites Grouped by Drug and Timepoint
grouped_table_sem = merge_table.groupby(['Drug', 'Timepoint']).sem()
# Convert to DataFrame
sem_metastatic_df = grouped_table_sem.reset_index()[['Drug', 'Timepoint', 'Metastatic Sites']]
# Preview DataFrame
sem_metastatic_df
```

```python
new_sem_metastatic_df = sem_metastatic_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
new_sem_metastatic_df
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(new_mean_metastatic_df.index, 
             new_mean_metastatic_df['Capomulin'], 
             yerr=new_sem_metastatic_df['Capomulin'], 
             marker='o', color='r', linewidth=0.5)

plt.errorbar(new_mean_metastatic_df.index, 
             new_mean_metastatic_df['Infubinol'], 
             yerr=new_sem_metastatic_df['Infubinol'], 
             marker='o', color='b', linewidth=0.5)

plt.errorbar(new_mean_metastatic_df.index, 
             new_mean_metastatic_df['Ketapril'], 
             yerr=new_sem_metastatic_df['Ketapril'], 
             marker='o', color='g', linewidth=0.5)

plt.errorbar(new_mean_metastatic_df.index, 
             new_mean_metastatic_df['Placebo'], 
             yerr=new_sem_metastatic_df['Placebo'], 
             marker='o', color='k', linewidth=0.5)

Capomulin, = plt.plot(new_mean_metastatic_df.index, 
                     new_mean_metastatic_df['Capomulin'], 
                     marker='o', color='red', label='Capomulin')

Infubinol, = plt.plot(new_mean_metastatic_df.index, 
                     new_mean_metastatic_df['Infubinol'], 
                     marker='o', color='blue', label='Infubinol')

Ketapril, = plt.plot(new_mean_metastatic_df.index, 
                     new_mean_metastatic_df['Ketapril'], 
                     marker='o', color='green', label='Ketapril')

Placebo, = plt.plot(new_mean_metastatic_df.index, 
                     new_mean_metastatic_df['Placebo'], 
                     marker='o', color='black', label='Placebo')


plt.xlabel('Treatment Duration (Days)')
plt.ylabel('Met. Sites')
plt.title('Metastatic Spread During Treatment')
plt.grid(axis='y')
plt.legend(loc='upper left')

plt.savefig('../Images/spread.png')
plt.show()
```

```python
# Store the Mouse Count 
grouped_table_c = merge_table.groupby(['Drug', 'Timepoint']).count()
count_mouse_df = grouped_table_c.reset_index()[['Drug', 'Timepoint', 'Mouse ID']]
count_mouse_df
```
